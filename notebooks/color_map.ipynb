```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Our color space
ANSI = {
    '40': (0, 0, 0),    '41': (170, 0, 0),    '42': (0, 170, 0),    '43': (170, 85, 0),
    '44': (0, 0, 170),  '45': (170, 0, 170),  '46': (0, 170, 170),  '47': (170, 170, 170),
    '100': (85, 85, 85),  '101': (255, 85, 85),  '102': (85, 255, 85),  '103': (255, 255, 85),
    '104': (85, 85, 255),  '105': (255, 85, 255),  '106': (85, 255, 255),  '107': (255, 255, 255),
}

C = np.array(list(ANSI.values()))
C = C/255

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')

for color in C:
    ax.scatter(color[0], color[1], color[2], c=[color], depthshade=False)

ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
ax.set_xlim(0, 1.25)
ax.set_ylim(0, 1.25)
ax.set_zlim(0, 1.25)
# ax.view_init(elev=20, azim=30)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.tight_layout(pad=10.0)
plt.show()
```

Euclidean Distance in 3D

$$ d = \sqrt{\left(x_2 - x_1\right)^2 + \left(y_2 - y_1\right)^2 + \left(z_2 - z_1\right)^2} $$

or

$$ d = \sqrt{\sum_{i=0}^n{ \left( x_{2i}-x_{1i} \right)^2 }} $$

Where:
* $d$ is Euclidean Distance
* $(x1, y1, z1)$ is Coordinate of the first point
* $(x2, y2, z2)$ is Coordinate of the second point

```python
# Search color
SC = (125, 10, 64)

fig = plt.figure(figsize=(1, 1))
plt.plot(0,0, marker='d', color=np.array(SC)/255)
plt.show()
```

```python
import math

def color_search(c1: tuple):
    distances = []
    for k,i in ANSI.items():
        distances.append(
            {
                "d": math.sqrt( (i[0] - c1[0])**2 + (i[1] - c1[1])**2 + (i[2] - c1[2])**2 ),
                "k": k
            }
        )
    distances.sort(key=lambda x: x["d"])
    return distances[0]["k"]

#######################################

close = color_search(SC)
print(ANSI[close])

fig = plt.figure(figsize=(1, 1))
plt.plot(0,0,marker='d',color=np.array(ANSI[close])/255)
plt.show()
```

```python
SC = (0, 100, 163)
close = color_search(SC)
print(ANSI[close])

fig = plt.figure(figsize=(1, 1))
plt.plot(0,0,marker='d',color=np.array(ANSI[close])/255)
plt.show()
```
